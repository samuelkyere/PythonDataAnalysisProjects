## Create Target and Predicted Variables

We use independent variables ('Time_on_Website','Time_on_App','Length_of_Membership','Avg_Session_Length'), as our predictors, X

```python
X = fyc[['Time_on_Website','Time_on_App','Length_of_Membership','Avg_Session_Length']
        ] #predictors
X
```

We use our dependent variable 'Yearly_Amount_Spent'as our target, y

```python
y = fyc['Yearly_Amount_Spent']  #target
y
```

 # We split our data in four types, our predictor x, into X_train, X_test; and our target values Y, into two Y_test and Y_train. 
 # We use test size of %25 to predict our X. And a random stae value of 85.
 # We do our prediction using linear regression model

```python
X_train,X_test, Y_train, Y_test = train_test_split(X, y, test_size= 0.25, random_state = 85)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
prediction = model.predict(X_test)
prediction
```

## Calculation of mean squared error on our dataset

 # We compare the actual 'yearly amount spent' with predicted 'yearly amount spent', and the deviations of values, which do not fit in the line of best fit are squared, the average calculated and the square root which gives us the mean squared value

```python
MSE = mean_squared_error(Y_test, prediction)
MSE
```

```python
fig, ax = plt.subplots()
ax.scatter( Y_test, prediction)
ax.plot([y.min(), y.max()], [y.min(), y.max()], 'r--', lw=2)
ax.set_xlabel('True Values')
ax.set_ylabel('Predicted')
plt.show()
```
